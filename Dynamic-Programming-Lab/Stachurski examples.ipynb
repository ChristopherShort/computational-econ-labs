### Example: A fisherman's problem

We wish to model the behavior of Colonel Kurtz, who can be found on a small island in the Nung River living on catfish.  Local catfish bite only at dawn, and the Colonel bags a randome number of fish $W \in [0, \dots, B]\ $ where $W \sim \phi$. Catfish spoil quickly in the tropics, lasting only if refrigerated. The Colonel has limited refrigeration facilities: a maximum of $M$ fish can fit into his freezer.  Let $X_t$ be the stock of fish at noon on day $t$, from which $C_t$ is to be consumed. The remainder $R_t = X_t - C_t$ is frozen.  Following the next morning's fishing trip, the state is $X_{t+1} = R_{t} + W_{t+1}$. Colonel Kurtz has discount factor $0 \lt \beta \lt 1\ $ and CRRA preferences over consumption. Upon observing state $X_t$, assume that Colonel Kurtz saves a quantity $R_t = \sigma(X_{t})\ $ (i.e., he follows a time-invariant storage policy). 

```python
# define utility parameters...
theta = 0.75
beta = 0.95 

# upper bound on size of fish catch
B = 50 

# storage capacity
M = 25

# define the state and shock space
S = range(B + M + 1)
Z = range(B + 1)        

# define some useful functions...
def d_infty(v, w):
    """d_infty metric"""
    return max(abs(v[x] - w[x]) for x in S)

def U(c):
    """CRRA Utility function"""
    return (c**(1 - theta) - 1) / (1 - theta)

def q(z):
    """Probability mass function, uniform distribution"""
    return 1.0 / len(Z) if 0 <= z <= B else 0

def gamma(x):
    """The correspondence of feasible actions"""
    return range(min(x, M) + 1)

def T(v):
    """An implementation of the Bellman Operator.
    
    Parameters:
        
        v, a sequence representing a function defined on S
    
    Returns:
        
        Tv, a list

    """
    Tv = []
    for x in S:
        # Compute the value of the objective function for each
        # a in Gamma(x), and store result in vals
        vals = []
        for a in gamma(x):
            y = U(x - a) + beta * sum(v[a + z] * q(z) for z in Z)
            vals.append(y)
            # Store the maximum reward for this x in the list Tv
        Tv.append(max(vals))
    return Tv

def greedy(v):
    """Computes the optimal (i.e., greedy policy) for the 
    value function v.

    """
    g = []
    for x in S:
        runningmax = -1
        for a in gamma(x):
            y = U(x - a) + beta * sum(v[a + z] * q(z) for z in Z)
            if y > runningmax:
                runningmax = y
                maximizer = a
        g.append(maximizer)
    return g
```

Here is a simple implementation of the value iteration algorithm for solving Colonel Kurtz' decision problem:

```python
# initial guess for value function
old_V = [U(x) for x in S]

# specify some tolerance...\n",
tol = 0.01 * (1 - beta)

#... and initialize dist
dist   = 1.0
n_iter = 0

# Value iteration algorithm
while dist > tol:
    new_V = T(old_V)
    dist  = d_infty(new_V, old_V)
    if n_iter % 10 == 0:
        print "After", n_iter, "iterations, the distance is",  dist
    old_V = new_V
    n_iter += 1
          
print "Convergence after", n_iter, "iterations!"
     
```

```python
def value_of_policy(sigma):
    """Computes the value of following policy sigma."""

    # Set up the stochastic kernel p_sigma as a 2-D array
    N = len(S)
    p_sigma = np.array([[q(y - sigma[x]) for y in S] for x in S])

    # Create the Markov operator M_sigma
    M_sigma = lambda h: np.dot(p_sigma, h)

    # Set up the function r_sigma as an array
    r_sigma = np.array([U(x - sigma[x]) for x in S])

    # Re-shape r_sigma into a column vector
    r_sigma = r_sigma.reshape((N,1))

    # Initialize v_sigma to zero
    v_sigma = np.zeros((N,1))

    # Initialize the discount factor to one
    discount = 1

    for i in xrange(100):
        v_sigma = v_sigma + beta**i * r_sigma
        r_sigma = M_sigma(r_sigma)

    return v_sigma

##### Main loop implements policy iteration #####

# new figure
plt.figure(figsize=(8,6))

# guess the current policy
current = np.zeros(len(S))

# initialize a counter
n_iter = 0

while 1:
    plt.plot(current, 'go--', alpha=0.25)
    v = value_of_policy(current)
    new = np.array(greedy(v))
    e = current - new
    n_iter += 1
    if all(e == 0):
        break
    current = new

# optimal policy is... 
g = new
plt.plot(g, 'go--')

# axes, labels, etc
plt.xlabel("Stock of fish, $X_t$", fontsize=15)
plt.ylabel("$\sigma^*(X_{t})$", rotation="horizontal", fontsize=15)
plt.ylim(0, M + 1)
plt.title("Policy iterates for Colonel Kurtz", fontsize=15)

print 'After', n_iter, 'interations, the optimal policy is:'
print list(g)

plt.show()
```

```python
import numpy as np
from scipy.optimize import minimize_scalar, fminbound
from scipy.stats import lognorm
from scipy.spatial.distance import chebyshev
from stepfun import StepFun

theta = 0.5
alpha = 0.33
beta  = 0.95
sigma = 1.0

def u(c): 
    return 1 - np.exp(- theta * c)  # Utility

# Cumulative distribution (cdf) of lognormal shock
L = lognorm(1, 0, sigma)
G = L.cdf   

def dist(c):
    """
    Creates and returns (as a function) the cdf of c * W, where W is 
    lognormal and c is a constant.
    """
    if c == 0.0:
        return np.vectorize(lambda x: 0 if x < 0 else 1)
    return lambda x: G(x / c)

# Next we make a grid.  The grid is constructed in a nonlinear way, 
# so that there are more points near the bottom of the grid.
# The reason is that the iterates have more curvature at the bottom,
# so we need more data there to approximate well.
gridmax, gridsize = 8, 150
grid = np.linspace(0, gridmax**1e-1, gridsize)**10

#def maximum(h, a, b):
#    "Computes the max of h on the interval [a, b]."
#    return h(fminbound(lambda x: -h(x), a, b))  

def maximum(h, bounds=None, bracket=None, method='Bounded'):
    """Finds the maximum value of the function h. Note that we find 
    the maximum value of the function h by minimizing -h. Wraps SciPy's 
    minimize_scalar interface.

    Arguments:
    
        h:       scalar function to be maximized.
        bounds:  tuple containing lower and upper bounds for use with
                 'Bounded' method.
        bracket: 
        method:  One of either 'Bounded', 'Golden', 'Brent'
    
    Function returns a tuple containing the maximum value of h and its 
    maximizer (i.e., the optimal value of the control).
    
    """
    result = minimize_scalar(lambda x: -h(x), bracket=bracket, bounds=bounds, 
                             method=method)
    value, control = -result.fun, result.x
    
    return (value, control)

def bellman(w):
    """
    The approximate Bellman operator.

     Parameters: 
     
        * w is an instance of StepFun with w.X = grid

    Returns: A new instance of StepFun 
    """
    Tw = np.empty(gridsize)
    for i, y in enumerate(grid):
        h = lambda k: u(y - k) + beta * w.expectation(dist(k**alpha))
        Tw[i] = maximum(h, bounds=(0, y))[0]
    return StepFun(grid, Tw)
```

```python
##### Value function iteration #####

# initial guess for value function
v   = StepFun(grid, u(grid))

# Error tolerance
tol = 0.01 * (1 - beta)      
num_iter = 0

while True:
    new_v = bellman(v)
    error  = chebyshev(new_v.Y, v.Y)
    num_iter += 1
    if error < tol:
        break
    else:
        # plot every 10th iterate
        if num_iter % 10 == 0:
            plt.plot(grid, v.Y, 'b-', alpha=0.25)
        # print convergence progress
        if num_iter % 1 == 0:
            plt.plot(grid, v.Y, 'b-')
            print "After", num_iter, "iterations, the current error is", error
        v = new_v

plt.show()
```
