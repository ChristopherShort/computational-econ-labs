This notebook solves and simulates an infinite horizon stochastic growth model.  As was the case in the deterministic version, there are two exogenous sources of growth in this model. The size of the labor force grows deterministically at rate $n$ per period:

$$ L_{t+1} = (1 + n)L_t \tag{1}$$

We will model technology using two different stochastic processes.  First, we will look at persistent fluctuations around a deterministic trend growth rate of $g$ per period:

$$
\begin{align}
A_{t+1} &= (1 + g)\left(\frac{z_{t+1}}{z_t}\right)A_t \\\
ln\ z_{t+1} &= \rho ln\ z_t + \epsilon_{t+1},\ |\rho|\lt1,\ \epsilon\sim\mathcal{N}(0,\sigma) \tag{2}
\end{align}
$$

and then, a unit root process with an expected growth rate of $g$ per period:

$$
\begin{align}
A_{t+1} &= z_{t+1}A_t \\\
ln\ z_{t+1} &= \rho ln\ (1 + g) + (1 - \rho) ln\ z_t + \epsilon_{t+1},\ |\rho|\lt1,\ \epsilon\sim\mathcal{N}(0,\sigma) \tag{3}
\end{align}
$$

In order to work with a stationary model we will work with per effective worker variables: the rate of consumption per effective worker is denoted by $c$, and $y=f(k)=k^{\alpha}\ $ is output per effective worker when the capital stock per effecive worker is $k$. Starting from this setup, the most natural state variable is $k\ $ which, if technology is modelled as persistant fluctuations around a deterministic trend, evolves according to:

$$k_{t+1} = \frac{1}{(1+g)(1+n)\left(\frac{z_{t+1}}{z_t}\right)}\left[(1 - \delta)k_t + k^{\alpha}_t - c_t\right]\tag{4}$$

If technology is modelled as a unit root process, then $k$ evolves according to:

$$k_{t+1} = \frac{1}{(1+n)z_{t+1}}\left[(1 - \delta)k_t + k^{\alpha}_t - c_t\right]\tag{5}$$

Assuming time separable CRRA preferences with discount factor $0 \lt \beta \lt 1\ $, the discrete-time optimal growth model becomes

$$ \max_{c_t}\ E\left\\{\sum_{t=0}^{\infty} [\beta(1+g)^{1-\theta}(1+n)]^t \frac{c_t^{1-\theta}}{1-\theta}\right\\} \tag{6}$$

subject to either equation 4 or 5 (depending on how technology is modelled) where $k_0$ is given. 

The first-order conditions for the optimal path are $c_t^{-\theta} = \beta(1+g)^{-\theta}E_t\\{c_{t+1}^{-\theta}(1 + \alpha k^{\alpha-1}_{t+1} - \delta)\\}\ \ \ $ for all $t$. The dynamic programming version of equation 6 is

$$ V(k_t) = \max_{0 \le c_t \le (1 - \delta)k_t + k^{\alpha}_t}\ \frac{c_t^{1-\theta}}{1-\theta} + \beta(1+g)^{1-\theta}(1+n) E_t\left\\{V\left(\frac{1}{(1+g)(1+n)}\left[(1 - \delta)k_t + k^{\alpha}_t - c_t\right]\right)\right\\} \tag{7}$$

Note that the feasibility set for choices of consumption per effective worker does not require $c_t \le f(k_t)$, but instead requires that $c_t \le (1-\delta)k_t + f(k_t)\ \ $. 

The optimal policy function, also called the optimal policy law, $c(k_t)$ obeys the first-order conditions

$$0 = c(k_t)^{-\theta} - \beta(1+g)^{-\theta} E_t\left\\{V'\left(\frac{1}{(1+g)(1+n)}\left[(1 - \delta)k + k^{\alpha} - c(k_t)\right]\right)\right\\} \tag{8}$$

The envelope theorem applied to 12.1.13 yields 

$$V'(k_t)=\beta(1+g)^{-\theta}(1+n) V'\left(\frac{1}{(1+g)(1+n)}\left[(1-\delta)k_t + k^{\alpha}_t - c_t\right]\right)[(1 -\delta) + \alpha k^{\alpha-1}_t]$$

which together with 12.1.14 implies that 

$$0 = c(k_t)^{-\theta}[(1 -\delta) + \alpha k^{\alpha-1}_t] - V'(k_t) \tag{12.1.15}$$

Thus the solution to 12.1.13 is a policy function $c(k_t)$ and a value function $V(k_t)$ that satisfy 12.1.15 and 

$$V(k_t) = \frac{c(k_t)^{1-\theta}}{1-\theta} + \beta(1+g)^{1-\theta}(1+n) V\left(\frac{1}{(1+g)(1+n)}\left[(1 - \delta)k_t + k^{\alpha}_t - c(k_t)\right]\right) \tag{12.1.16} $$

Note that 12.1.16 defines the value of an arbitrary policy function $c(k_t)\ $, not just for the optimal policy. The pair 12.1.15 and 12.1.16 combine the definition of the value function with the first-order optimality condition. 

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from scipy import interp, stats
from scipy.interpolate import UnivariateSpline, RectBivariateSpline, LinearNDInterpolator
from scipy.spatial.distance import chebyshev
from scipy.optimize import minimize_scalar, fminbound

import srs
from lininterp import LinInterp
from stepfun import StepFun
from pyeconomics.utility import lifetimeUtility
```

```python
def u(c):
    """Agent has CRRA preferences where c is consumption per
    effective worker.

    In general, the flow of utility can depend on the state, but
    in this simple problem, the state variable does not enter 
    into the utility!
    
    """
    if theta != 1:
        return (c**(1 - theta)) / (1 - theta)
    else:
        return np.log(c)
    
def capital(k, z, c):
    """Equation of motion for capital per effective worker."""
    kplus = (1 / ((1 + g) * (1 + n) * z)) * ((1 - delta) * k + k**alpha - c)
    return kplus
    
def k_star():
    """Steady state value of capital per effective worker."""
    kstar = (alpha * beta / ((1 + g)**theta - beta * (1 - delta)))**(1 / (1 - alpha))
    return kstar

def c_star():
    """Steady state value of consumption per effective worker."""
    cstar = k_star()**alpha - ((1 + g) * (1 + n) - (1 - delta)) * k_star()
    return cstar

def gamma(k):
    """Bounds on the feasible set of choices of consumption 
    per effective worker.

    """
    lower = 0.0
    upper = (1 - delta) * k + k**alpha
    return (lower, upper)
```

```python
def maximum(h, bounds=None, bracket=None, method='Bounded'):
    """Finds the maximum value of the function h. Note that we find 
    the maximum value of the function h by minimizing -h. Wraps SciPy's 
    minimize_scalar interface.

    Arguments:
    
        h:       scalar function to be maximized.
        bounds:  tuple containing lower and upper bounds for use with
                 'Bounded' method.
        bracket: 
        method:  One of either 'Bounded', 'Golden', 'Brent'
    
    Function returns a tuple containing the maximum value of h and its 
    maximizer (i.e., the optimal value of the control).
    
    """
    result = minimize_scalar(lambda x: -h(x), bracket=bracket, bounds=bounds, 
                             method=method)
    value, control = -result.fun, result.x
    
    return (value, control)
```

```python
def nanmean(a):
    """Convenience function computing the mean of an array
    including NaNs.

    Arugments:
    
        a: array-like.

    """
    return np.mean(np.ma.MaskedArray(a, np.isnan(a)))
```

```python
def discreteBellman2D(w):
    """An implementation of the Bellman operator. Used to perform
    value function iteration over a discretized grid of point in 2D.
    
    Arguments:
        
        w: an array representing a function defined over the 
           discrete grid of possible states.
    
    Returns: A list containing
        
        Tw:            an array of values representing the 
                       next value function iterate.
        greedy_policy: an array of values representing the 
                       next policy function iterate.

    """
    # Compute next value function iterate (note that w is being broadcast!)
    Tw = np.nanmax(u(kStates, zStates, kControls) + discount * w.reshape(Nk,1), axis=0)
    
    # Compute next policy function iterate (note that w is being broadcast!)
    #greedy_indices = np.nanargmax(u(kStates, kControls) + discount * w.reshape(Nk,1), axis=0) 
    #greedy_policy  = kGrid[greedy_indices]
    
    return [Tw, greedy_policy]
```

```python
def randomBellman(w):
    """The randomized Bellman operator, R, from Stachurski and Pal (2012)"""
    Tw = np.empty((Nk,))
    greedy_policy = np.empty((Nk,))
    
    for i, k in enumerate(kGrid):
        # current value function (note that expectations are computing by taking the average!)
        h = lambda c: u(c) + discount * nanmean(w(capital(k, Z, c)))
        # find the value of the control that maximizes the current value function
        val, pol = maximum(h, gamma(k))
        # update the value function with maximum value of h
        Tw[i] = val
        # update the policy function with the maximizer of h
        greedy_policy[i] = pol

    return (UnivariateSpline(kGrid, Tw, k=1, s=0), UnivariateSpline(kGrid, greedy_policy, k=1, s=0))
```

```python
# define model parameters
theta = 2.5
beta  = 0.95
g     = 0.02 
n     = 0.02 
delta = 0.10
alpha = 0.33
rho   = 0.0
sigma = 0.01

# define the discount factor
discount = beta * (1 + g)**(1 - theta) * (1 + n)
```

```python
# Draws of the shock
T = 1000
Z = np.random.lognormal(0.0, sigma, T)

# grid of capital per effective worker and technology shocks (i.e., state variables!)
Nk = 250
kMax = (1 / ((1 + g) * (1 + n) - (1 - delta)))**(1 / (1 - alpha))
kMin = kMax / Nk
kGrid = np.linspace(kMin, kMax, Nk)
```

```python
# Value function iteration algorithm
def initial_v(k):
    """Being smart about our initial guess can save substantial 
    amount of computational time!

    """
    return u(c_star()) / (1 - discount)
```

```python
##### Plot setup #####

# create a new figure instance (we are going to plot value and policy iterates as we go!)
fig = plt.figure(figsize=(12,6))

# grid for plotting (different from state variable grid)
plot_grid = np.linspace(kMin, kMax, 1000)

# create two subplots, one for value function and one for policy function
ax1 = fig.add_subplot(121)
ax1.plot(plot_grid, [initial_v(k) for k in plot_grid], 'b--', alpha=1.0, label='$V_0(k_t)$')
ax1.set_xlabel('$k_t$', fontsize=15)
ax1.set_ylabel('$V^*(k_t)$', fontsize=15, rotation='horizontal')
ax1.set_title('Optimal value function', weight='bold')

ax2 = fig.add_subplot(122)
ax2.plot(plot_grid, [c_star() for k in plot_grid], 'g--', alpha=1.0, label='$c_0(k_t)$')
ax2.set_xlabel('$k_t$', fontsize=15)
ax2.set_ylabel('$c(k_t)$', fontsize=15, rotation='horizontal')
ax2.set_title('Optimal consumption policy', weight='bold')

##### Fitted value function iteration algorithm #####

# initialize value function, and num_iter
current_v = initial_v
num_iter  = 0

# actual tolerance is a function of discount factor!
tol       = 0.01 * (1 - discount)

while True:
    next_v, next_pol = randomBellman(current_v)
    error = chebyshev(current_v(kGrid), next_v(kGrid))
    num_iter += 1
    if error < tol:
        final_v, final_pol = next_v, next_pol
        print "After", num_iter, "iterations, the final error is", error
        ax1.plot(plot_grid, final_v(plot_grid), 'b-', label='$V^*(k)$')
        ax2.plot(plot_grid, final_pol(plot_grid), 'g-', label='$c_{RVI}(k)$')
        break
    else:
        current_v, current_pol = next_v, next_pol
        # plot every 10th iterate
        if num_iter % 10 == 0:
            ax1.plot(plot_grid, current_v(plot_grid), 'b--', alpha=0.1)
            ax2.plot(plot_grid, current_pol(plot_grid), 'g--', alpha=0.1)
        # update on convergence 
        if num_iter % 50 == 0:
            print "After", num_iter, "iterations, the current error is", error

# add legends to the plot
ax1.legend(loc='best', frameon=False)
ax2.legend(loc='best', frameon=False)

fig.tight_layout()
#plt.savefig('Graphics/Stochastic-Ramsey/Optimal-value-policy-function-iterates-RVI.png')
plt.show()
```

```python
# store the optimal policy and value functions
ramsey_policyFunction = final_pol
ramsey_valueFunction  = final_v
```

```python
##### Compare the optimal, Ramsey vs rule of thumb, Solow policies #####
fig = plt.figure(figsize=(8,6))
ax1 = fig.add_subplot(111)

# plot the optimal policy function 
ax1.plot(plot_grid, ramsey_policyFunction(plot_grid), 'r-', label='$c_{RVI}(k)$')

# plot the Solow rule of thumb policy function
s = (k_star()**alpha - c_star()) / k_star()**alpha
ax1.plot(plot_grid, (1 - s) * plot_grid**alpha, 'g-', label='$c_{Solow}(k)$')

# demarcate the steady state
ax1.vlines(k_star(), 0, c_star(), linestyles='dashed', colors='k')
ax1.hlines(c_star(), 0, k_star(), linestyles='dashed', colors='k')

# axes, labels, title, legend, etc
#ax1.set_xlim(0, 3 * k_star())
ax1.set_xlabel('$k_t$', fontsize=15)
ax1.set_ylabel('$c(k_t)$', fontsize=15, rotation='horizontal')
ax1.set_title('Ramsey vs. Solow consumption policy functions', fontsize=15, weight='bold')
ax1.legend(loc='best', frameon=False)
plt.show()
```

```python
# Define the shock process
def eps():
    """Returns a draw from a lognormal distribution."""
    return np.random.lognormal(0, sigma, 1)
```

```python
# wrap the optimal consumption policy
c_ramsey = lambda k: ramsey_policyFunction(k)

# wrap the Solow consumption policy
c_solow = lambda k: (1 - s) * k**alpha

# Define the equation of motion for k using the "Ramsey" policy
F_ramsey = lambda k, z: capital(k, z, c_ramsey(k))

# Define the equation of motion for k using the "Solow" policy
F_solow = lambda k, z: capital(k, z, c_solow(k))
    
# create an instance of the SRS object
ramsey_capitalSRS = srs.SRS(F=F_ramsey, phi=eps)
solow_capitalSRS  = srs.SRS(F=F_solow, phi=eps)
```

### Policy iteration

```python
def LoT(policy, v):
    """Implementation of the L o T_sigma operator for approximating
    the value function for a specified policy using linear 
    interpolation.

    Arguments:

        policy: a callable consumption policy function.  As written 
                the L o T operator assumes that consumption is a 
                function of capital.
        v:      a callable value function.
    
    """
    vals = []
    
    for k in grid:
        Tw_y = u(policy(k)) + discount * np.mean(v(capital(k, Z, policy(k))))
        vals.append(Tw_y)
        
    return LinInterp(grid, vals)

def get_greedyPolicy(v):
    "Computes the optimal, greedy policy for the value function v."
    controls = []
    
    for k in grid:
        h = lambda c: u(c) + discount * np.mean(v(capital(k, Z, c)))
        controls.append(maximum(h, bounds=gamma(k))[1])
    
    return LinInterp(grid, controls)

def get_policyValue(policy, init_v, tol=1.5e-4, mesg=True):
    """Computes the value function for a specified policy.

    Arguments:

        policy: a callable policy function for which you seek the 
                corresponding policy function.
        init_v: your best guess of the value function you seek.
        tol:    error tolerance.

    """
    # start from the initial guess
    current_v = init_v
    n_iter = 0
    
    while True:
        new_v = LoT(policy, current_v)
        error = chebyshev(new_v(grid), current_v(grid))
        n_iter += 1
        # Check convergence
        if error < tol:
            final_v = new_v
            break
        # print progress 
        if mesg == True and n_iter % 50 == 0:
            print "After", n_iter, "iterations, the error is", error
        current_v = new_v
    
    return final_v
```

```python
solow_valueFunction = get_policyValue(c_solow, np.log)
```

```python
def proportionateBonus(k):
    """State dependent proportionate welfare gains from following optimal, Ramsey
    policy versus the rule of thumb Solow alternative.

    """
    if theta != 1.0:
        return ((ramsey_valueFunction(k) - discount * np.mean(solow_valueFunction(F_solow(k, Z)))) / u(c_solow(k)))**(1 / (1 - theta))
    else:
        return (ramsey_valueFunction(k) - discount * np.mean(solow_valueFunction(F_solow(k, Z)))) / u(c_solow(k))
```

```python
##### Fitted policy iteration algorithm #####

# Use the rule of thumb, Solow policy as our initial guess
initial_policyFunction = c_solow

# initialize a counter and the current value and policy functions
num_iter = 0
current_policyFunction = initial_policyFunction
current_valueFunction = np.log

# choose a convergence tolerance
tol = 0.01 * (1 - discount)

while True:
    # compute the value function implied by the current policy
    next_valueFunction = get_policyValue(current_policyFunction, init_v=current_valueFunction, mesg=False)
    # compute the greedy policy for this value function
    next_policyFunction = get_greedyPolicy(next_valueFunction)
    # check for convergence
    error = chebyshev(current_policyFunction(grid), next_policyFunction(grid))
    num_iter += 1
    if error < tol:
        final_valueFunction, final_policyFunction = next_valueFunction, next_policyFunction
        print "After", num_iter, "iterations, the final error is", error
        break
    else:
        current_valueFunction, current_policyFunction = next_valueFunction, next_policyFunction
        # update on convergence 
        if num_iter % 1 == 0:
            print "After", num_iter, "iterations, the current error is", error
```

```python
# parameters of the model
theta = 0.5
alpha = 0.33
beta  = 0.95
sigma = 1.0       

def u(c): 
    """Constant absolute risk aversion preferences.""" 
    return 1 - np.exp(- theta * c)
```

```python
# Cumulative distribution (cdf) of lognormal shock
L = stats.lognorm(1, loc=0, scale=sigma)
G = L.cdf   

def dist(c):
    """
    Creates and returns (as a function) the cdf of c * W, where W is 
    lognormal and c is a constant.
    """
    if c == 0.0:
        return np.vectorize(lambda x: 0 if x < 0 else 1)
    return lambda x: G(x / c)
```

```python
gridmax, N = 8, 500
grid = np.linspace(0, gridmax**1e-1, N)**10
```

```python
def bellman(v):
    """An approximate Bellman operator.

    Parameters: 
     
        w: an instance of StepFun with w.X = grid

    Returns: A list containing...

        1. A new instance of StepFun representing the 
           value function iterate.
        2. The greedy policy for that value function.

    """
    # storage container for new Value function iterate
    Tv            = np.empty(N)
    greedy_policy = np.empty(N)
    
    for i, y in enumerate(grid):
        h = lambda k: u(y - k) + beta * v.expectation(dist(k**alpha))
        Tv[i], greedy_policy[i] = maximum(h, bounds=(0, y))
    
    return [StepFun(grid, Tv), greedy_policy]
```

```python
##### Value function iteration #####
plt.figure(figsize=(8,6))

# initial guess for value function
v   = StepFun(grid, u(grid))

# Error tolerance
tol = 0.01 * (1 - beta)      
num_iter = 0

while True:
    new_v, new_pol = bellman(v)
    error  = chebyshev(new_v.Y, v.Y)
    num_iter += 1
    if error < tol:
        final_v, final_pol = new_v, new_pol
        plt.plot(grid, final_v.Y, 'b-', label='$V^*(y_t)$')
        break
    else:
        # plot every 10th iterate
        if num_iter % 10 == 0:
            plt.plot(grid, v.Y, 'b-', alpha=0.25)
        # print convergence progress
        if num_iter % 50 == 0:
            print "After", num_iter, "iterations, the current error is", error
        v = new_v
     
plt.ylabel('$V(y_t)$', rotation='horizontal', fontsize=15)
plt.xlabel('$y_t$', fontsize=15)
plt.title('Optimal value function using step-function approximation', weight='bold', fontsize=15)          
plt.legend(loc='best', frameon=False)
plt.show()
```

```python
class PWC:

    def __init__(self, xgrid, ygrid, vals):
        self.xgrid, self.ygrid, self.vals = xgrid, ygrid, vals

    def __call__(self, xpoints, ypoints):
        xindices = self.xgrid.searchsorted(xpoints, side='right') - 1
        yindices = self.ygrid.searchsorted(ypoints, side='right') - 1
        return self.vals[yindices, xindices]


######################### Primitives ######################################

RHO = 0.9     # Discount rate
LAMBDA = 0.7  # Carryover depreciation parameter
THETA = 0.3   # Parameter in the shock process
ALPHA = 0.2   # Utility parameter

# Innovation is a + b W, where W is beta(5,5)
a = 1; b = 2
B = stats.beta(5, 5, loc=a, scale=b)
G = B.cdf    # G is the cdf of this distribution
W = B.rvs(1000)
TW = (1 - THETA) * W

########################### Grid #########################################

# Next we set out a grid on the state space S x H, where S is values for
# the first state variable (supply), and H is values for the shock.

SHOCK_UB = a + b                   # Upper bound of the innovation
SHOCK_LB = a                       # Lower bound of the innovation
J = 250                            # Number of elements in S-grid
K = 20                             # Number of elements in H-grid
S_UPPER = SHOCK_UB / (1 - LAMBDA)  # Upper bound of S
S_LOWER = SHOCK_LB                 # Lower bound of S
S = np.linspace(S_LOWER, S_UPPER, J)
H = np.linspace(SHOCK_LB, SHOCK_UB, K)


def bellman(v):
    """An implementation of the Bellman operator"""
    
    Tv = np.empty((K, J))
    greedy_policy = np.empty((K, J))
    for j in xrange(J):
        for k in xrange(K):
            nh = THETA * H[k] + TW
            ob = lambda q: (S[j] - q)**ALPHA + RHO * np.mean(v(LAMBDA * q + nh, nh))
            val, pol = maximum(ob, bounds=(0, S[j]))
            Tv[k, j] = val
            greedy_policy[k, j] = pol
            
    return [PWC(S, H, Tv), PWC(S, H, greedy_policy)]
```

```python
# initial guess for value function is flow utility from consuming storage
initial_vals = np.repeat(S[np.newaxis,:]**alpha, H.shape[0], axis=0)
v = PWC(S, H, initial_vals)

# Error tolerance
tol = 0.01 * (1 - beta)      
num_iter = 0

while True:
    new_v, new_pol = bellman(v)
    error  = np.max(np.abs(new_v.vals - v.vals))
    num_iter += 1
    if error < tol:
        final_v, final_pol = new_v, new_pol
        print "After", num_iter, "iterations, the final error is", error
        break
    else:
        # print convergence progress
        if num_iter % 10 == 0:
            print "After", num_iter, "iterations, the current error is", error
        v = new_v
```
