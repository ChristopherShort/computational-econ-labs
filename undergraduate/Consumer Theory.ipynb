```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
from mpl_toolkits.mplot3d import Axes3D
```

```python
def perfect_substitutes(x1, x2, w1=1, w2=1):
    """Consumer preferences when goods X1, X2 are perfect substitutes.

    Arguments:

        x1: quantity of good 1 consumed.
        x2: quantity of good 2 consumed.
        w1: utility weight.
        w2: utility weight.

    """
    return w1 * X1 + w2 * X2

def convex_preferences(x1, x2, w1=1, w2=1):
    """Example of standard convex preferences.

    Arguments:

        x1: quantity of good 1 consumed.
        x2: quantity of good 2 consumed.
        w1: utility weight.
        w2: utility weight.

    """
    return X1**w1 * X2**w2

def perfect_complements(x1, x2, w1=1, w2=1):
    """Consumer preferences when goods X1, X2 are perfect complements.

    Arguments:

        x1: quantity of good 1 consumed.
        x2: quantity of good 2 consumed.
        w1: utility weight.
        w2: utility weight.

    """
    return np.minimum(x1, x2)
```

```python
# create a grid of x1, x2 bundles at which to evaluate utility
x1 = np.linspace(0, 10, 100)
x2 = np.linspace(0, 10, 100)

X1, X2 = np.meshgrid(x1, x2)
```

```python
########## 3D plots of utility surfaces ##########

fig = plt.figure(figsize=(18,6))
cmap = mpl.cm.cool_r

##### Perfect substitutes #####
ax1 = fig.add_subplot(131, projection='3d', azim=-130, elev=30)

U = perfect_substitutes(X1, X2)
ax1.plot_surface(X1, X2, U, rstride=1, cstride=1, cmap=cmap, 
                linewidth=0, vmin=0, vmax=20, antialiased=False)

# axes, labels, title, etc.
ax1.set_xlabel('Good 1')
ax1.set_ylabel('Good 2')
ax1.set_title('Perfect Substitutes', weight='bold')

##### Standard convex preferences #####
ax2 = fig.add_subplot(132, projection='3d', azim=-130, elev=30)

U = 2 * convex_preferences(X1, X2, w1=0.5, w2 = 0.5)
ax2.plot_surface(X1, X2, U, rstride=1, cstride=1, cmap=cmap, 
                linewidth=0, vmin=0, vmax=20, antialiased=False)

# axes, labels, title, etc.
ax2.set_xlabel('Good 1')
ax2.set_ylabel('Good 2')
ax2.set_zlim(0, 20)
ax2.set_title('Standard convex preferences', weight='bold')

##### Perfect complements #####
ax3 = fig.add_subplot(133, projection='3d', azim=-130, elev=30)

U = 2 * perfect_complements(X1, X2)
ax3.plot_surface(X1, X2, U, rstride=1, cstride=1, cmap=cmap, 
                linewidth=0, vmin=0, vmax=20, antialiased=False)

# axes, labels, title, etc.
ax3.set_xlabel('Good 1')
ax3.set_ylabel('Good 2')
ax3.set_zlim(0, 20)
ax3.set_title('Perfect complements', weight='bold')

plt.show()
```
