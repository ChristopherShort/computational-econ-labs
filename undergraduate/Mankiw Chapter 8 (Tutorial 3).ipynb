```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint
```

```python
# dictionary of parameter values for a solow model (note technology growth is suppressed!)
solow_params = {'alpha':0.33, 'delta':0.10, 'n':0.02, 's':0.33}
```

### Q1: Intensive production and over saving.  

Assume that production is a function of capital and labour. Further, assume that the rate of savings, depreciation, population growth are all constant, as described in Chapter 8’s version of the Solow Model.  Finally, assume that the production can be described by the Cobb-Douglas function:

$$Y = F(K,L) = K^{0.5} L^{0.5}$$

where $K$ is capital and $L$ is labor.

**Part a)** What is the per worker production function, $y=f(k)\ $?

$$ y= \frac{Y}{L} = \frac{K^{0.5} L^{0.5}}{L} = K^{0.5}L^{-0.5} = \left(\frac{K}{L}\right)^{0.5} = k^{0.5} $$

```python
def output(k, t=0):
    """Output per worker as a function of capital per worker.

    Parameters:

        alpha: capital's share of income/output.

    """
    # extract the parameter values
    alpha = solow_params['alpha']
    
    return k**alpha
```

**Part b)** If the saving rate, $s\ $,  is 0.4, what are capital per worker, production per worker, and consumption per worker in the steady state?  (Note:  you need to set $\Delta k = 0$, to get an equation in $s$, $\delta$, $n$, and $k$, and then solve for $k$).

$$
\begin{align*}
0 &= sf(k) - (n + \delta)k \\\
sf(k) &= (n + \delta)k \\\
sk^{0.5} &= (n + \delta)k \\\
sk^{-0.5} &= (n + \delta) \\\
s\frac{1}{k^{0.5}} &= (n + \delta) \\\
k^{0.5} &= \frac{s}{n + \delta} \\\
k^* &= \left(\frac{s}{n + \delta}\right)^2 = \left(\frac{0.4}{n + \delta}\right)^2 \\\
y^* &= \left(\frac{s}{n + \delta}\right) = \left(\frac{0.4}{n + \delta}\right) \\\
c^* &= (1 - s)y^* = (1 - s)\left(\frac{s}{n + \delta}\right) = (1 - 0.4)\left(\frac{0.4}{n + \delta}\right) = \left(\frac{0.24}{n + \delta}\right)
\end{align*}
$$

```python
def capital(k, t=0):
    """Equation of motion for capital per worker for the Solow model."""
    # extract params
    delta = solow_params['delta']
    n     = solow_params['n']
    s     = solow_params['s']
    
    return  s * output(k, t) - (n + delta) * k

def k_star():
    """Steady state value of capital per worker for Solow model."""
    # extract params
    alpha = solow_params['alpha']
    delta = solow_params['delta']
    n     = solow_params['n']
    s     = solow_params['s']
    
    return (s / (n + delta))**(1 / (1 - alpha))

def y_star():
    """Steady state value of output per worker for Solow model."""
    # extract params
    alpha = solow_params['alpha']
    delta = solow_params['delta']
    n     = solow_params['n']
    s     = solow_params['s']
    
    return (s / (n + delta))**(alpha / (1 - alpha))

def c_star():
    """Steady state value of consumption per worker for Solow model."""
    # extract params
    alpha = solow_params['alpha']
    delta = solow_params['delta']
    n     = solow_params['n']
    s     = solow_params['s']
    
    return (1 - s) * (s / (n + delta))**(alpha / (1 - alpha))
```

```python
# grid of time points at which to compute the value of k
t = np.linspace(0, 100, 1000)

# use SciPy's built in ode solver to find the time path of k
k = odeint(capital, k_star() / 2, t)
```

```python
# plot the time path of capital per worker, k
plt.figure(figsize=(8,6))

# plot the time path of k
plt.plot(t, k)

# demarcate steady state value of k
plt.axhline(k_star(), color='k', linestyle='dashed', label='$k^*$')

# axes, labels, title, legend, etc
plt.xlabel('Time', fontsize=15)
plt.ylabel('$k_t$', fontsize=15, rotation='horizontal')
plt.title('Timepath of capital per worker\n' + r'$n=%0.3f,s=%.2f,\alpha=%.2f,\delta=%.2f$' \
         %(solow_params['n'], solow_params['s'], solow_params['alpha'], solow_params['delta']),
         weight='bold', fontsize=20)

plt.show()
```

**Part c)** Solve for steady-state capital per worker, production per worker, and consumption per worker with $s = 0.8\ $.

$$
\begin{align*}
k^* &= \left(\frac{s}{n + \delta}\right)^2 = \left(\frac{0.8}{n + \delta}\right)^2 \\\
y^* &= \left(\frac{s}{n + \delta}\right) = \left(\frac{0.8}{n + \delta}\right) \\\
c^* &= (1 - s)y^* = (1 - s)\left(\frac{s}{n + \delta}\right) = (1 - 0.8)\left(\frac{0.8}{n + \delta}\right) = \left(\frac{0.16}{n + \delta}\right)
\end{align*}
$$



**Part d)** Is it possible to save too much?  Why?

Yes, it is possible to save too much, in the sense that long-run consumption could actually be higher if you saved less.  Indeed, $s = 0.8\ $ appears to be too high a savings rate, since steady state consumption per person is definitely smaller with $s = 0.8\ $ than with $s = 0.4\ $.  Notice that we can conclude this even without knowing what the population growth rate or the depreciate rate for the economy are.

By the way, could we figure out what the actual “best” savings rate is, based on just the information given?  Yes, with a bit of deductive reasoning.  The golden rule level of capital is where $MPK = n + \delta\ $ :

$$0.5k^{-0.5} = MPK = n + \delta \implies k_{gold} = \left(\frac{0.5}{n + \delta}\right)^2$$

Thus the "golden rule" savings rate is $s=0.50\ $.

### Q2: Too much capital?  
Assume once again that production is given by $Y=F(K,L)=K^{0.5}L^{0.5}\ $.  First, write the production function in per person terms, $y=f(k)\ $.  Next, assume that the per person level of capital in the steady state is 4, the depreciation rate is 7% per year, and population growth is 3% per year.  Does this economy have “too much” or “too little” capital?  How do you know?

Again, if we are at the golden rule level of capital, then $MPK = n + \delta\ $.  Given the values of the parameters, and how much capital per person there is:

$$MPK = 0.5k^{-0.5} = 0.5(4)^{-0.5} = 0.5^2 = 0.25 > 0.07 + 0.03 = 0.1$$

Since $MPK > n + \delta\ $ we can conclude that there is “too little” capital in the economy.  With so little capital, the marginal productivity of this capital is quite high.

###Q4: Population shocks (Mankiw Chapter 8, Q1).

In the discussion of German and Japanese postwar growth, the text describes what happens when part of the capital stock is destroyed in a war. By contrast, suppose that a war does not affect the capital stock, but that casualties reduce the labour force.

**Part a)** What is the immediate impact on total output and on output per person?

The production function in the Solow growth model is $Y=F(K,L)=K^{\alpha}L^{1-\alpha}\ $, or expressed in terms of output per worker, $y=f(k)=k^{\alpha}$ . If a war reduces the labour force through casualties, then $L$ falls but $k=\frac{K}{L}$ rises. The production function tells us that total output falls because there are fewer workers. Output per worker increases, however, since each worker has more capital. 

**Part b)** Assuming that the saving rate is unchanged and that the economy was in a steady state before the war, what happens subsequently to output per worker in the postwar economy? Is the growth rate of output per worker after the war smaller or greater than normal?

The reduction in the labour force means that the capital stock per worker is higher after the war. Therefore, if the economy were in a steady state prior to the war, then after the war the economy has a capital stock that is higher than the steady state level. This is shown in the figure below as an increase in capital per worker from $k^*$ to $k_1$. As the economy returns to the steady state, the capital stock per worker falls from $k_1$ back to $k^*$, so output per worker also falls.

```python
# new figure
plt.figure(figsize=(8,6))

# grid of values for capital per worker
grid = np.linspace(0, 2 * k_star(), 1000)

# plot output, actual investment, and breakeven investment
plt.plot(grid, output(grid), 'r', label='$f(k)$')
plt.plot(grid, solow_params['s'] * output(grid), 'g', label='$sf(k)$')
plt.plot(grid, (solow_params['n'] + solow_params['delta'])* grid, 
         'b', label='$i_{break}$')

# axes, labels, title, legend, etc
plt.xlabel('$k_t$', fontsize=15)
plt.ylabel('$y_t, i_t, c_t$', fontsize=15, rotation='horizontal')
plt.title('Classic Solow diagram\n' + r'$n=%0.3f,s=%.2f,\alpha=%.2f,\delta=%.2f$' \
         %(solow_params['n'], solow_params['s'], solow_params['alpha'], solow_params['delta']),
         weight='bold', fontsize=20)

# demarcate the steady state values
plt.vlines(k_star(), 0, output(k_star()), linestyles='dashed', colors='k')
plt.hlines([solow_params['s'] * output(k_star()), output(k_star())], 
           0, k_star(), linestyles='dashed', colors='k')

# annotate the plot
plt.annotate('$k^*$', xy=(1.05 * k_star(), 0.05))
plt.annotate('$i^*$', xy=(0.2, 0.85 * solow_params['s'] * output(k_star())))
plt.annotate('$y^*$', xy=(0.1, 0.95 * output(k_star())))

plt.legend(loc='best', frameon=False)
plt.show()
```

###Q6: Population growth and returns to scale (Mankiw Chapter 8, Q8).

In the Solow model, population growth leads to steady-state growth in total output, but not in output per worker. Do you think this would still be true if the production function exhibited increasing or decreasing returns to scale? Explain.

If there are decreasing returns to labour and capital, then increasing both capital and labour by the same proportion increases output by less than this proportion. For example, if we double the amounts of capital and labour, then output less than doubles. This may happen if there is a fixed factor such as land in the production function, and it becomes scarce as the economy grows larger. Then population growth will increase total output but decrease output per worker, since each worker has less of the fixed factor to work with.

###Q7: Introducing unemployment (Mankiw Chapter 8, Q7).

Consider how unemployment would affect the Solow growth model. Suppose that output is produced according to the production function 

$$Y=F(K,L)=K^{\alpha}((1-u)L)^{1-\alpha}$$ 

where $K$ is capital, $L$ is the labour force, and $u$ is the natural rate of unemployment. The national saving rate is $s$, the labour force grows at rate $n$, and capital depreciates at rate $\delta$.


**Part a)** Express output per worker, $y = \frac{Y}{L}$, as a function of capital per worker, $k = \frac{K}{L}$, and the natural rate of unemployment, $u$.

$$ y = \frac{Y}{L}=(1-u)^{1-\alpha}K^{\alpha}L^{-\alpha}=(1-u)^{1-\alpha}\left(\frac{K}{L}\right)^{\alpha}=(1-u)^{1-\alpha}k^{\alpha}$$

Notice that unemployment reduces the amount of output per worker for any given capital–labour ratio because some of the workers are not producing anything.

```python
def capital(k, t=0):
    """Equation of motion for capital per worker for the Solow model."""
    # extract params
    delta = solow_params['delta']
    n     = solow_params['n']
    s     = solow_params['s']
    
    return s * output(k, t) - (n + delta) * k

def output(k, t=0):
    """Output per worker is a function of captial per worker.

    Parameters:
    
        alpha: capital's share of income/output
        u:     the natural rate of unemployment.

    """
    # extract params
    alpha = solow_params['alpha']
    u     = solow_params['u']

    return (1 - u)**(1 - alpha) * k**alpha
```

**Part b)** Write an equation that describes the steady state of this economy. Illustrate the steady state graphically.

The steady state is the level of capital per worker at which the increase in capital per worker from investment equals its decrease from depreciation and population growth.

$$
\begin{align}
s(1-u)^{1-\alpha}k^{\alpha} &= (n + \delta)k \\\
s(1-u)^{1-\alpha}k^{\alpha-1} &= (n + \delta) \\\
\frac{s(1-u)^{1-\alpha}}{k^{1-\alpha}} &= (n + \delta) \\\
k^{1-\alpha} =&\frac{s(1-u)^{1-\alpha}}{n + \delta} \\\
k^* =& (1 - u)\left(\frac{s}{n + \delta}\right)^{\frac{1}{1-\alpha}}
\end{align}
$$ 

Unemployment lowers the marginal product of capital per worker and, hence, acts like a negative technological shock that reduces the amount of capital the economy can maintain in steady state. The figure below shows this graphically: an increase in unemployment lowers output per worker which leads to lower actual investment and thus a lower the steady-state level of capital per worker.

```python
def k_star():
    """Steady state value of capital per worker for Solow model 
    with unemployment.

    """
    # extract params
    alpha = solow_params['alpha']
    delta = solow_params['delta']
    n     = solow_params['n']
    s     = solow_params['s']
    u     = solow_params['u']
    
    return (1 - u) * (s / (n + delta))**(1 / (1 - alpha))
    
```

```python
# add an unemployment rate to the dictionary of parameter values
solow_params['u'] = 0.10
```

```python
# what is the steady state value of k for these parameters
print "The steady state value of capital per worker is:", k_star()
```
