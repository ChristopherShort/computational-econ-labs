## 1.5 Quantitative implications

### Speed of convergence

In practice, we are interested not only in the *eventual* effects of some parameter change (such as a change in the savings rate), but also in how rapidly those effects occur.  In the case of the Solow model, we are often interested in the speed with which the economy converges to its long-run steady state. The speed with which the economy is converging toward its long run equilibrium at any point in time is given by the key equation of the model:

$$\dot{k} = sf(k(t)) - (n + g + \delta)k(t),\ k(0)=k_0$$

where $k_0$ is some given initial condition. The parameters $s, n, g, \delta$ and the intensive form of the production function are all known. If we knew the functional form for $k(t)$, then we could just make all the necessary substitutions to compute $\dot{k}$ directly. Unfortunately, in general, we will not know the functional form of $k(t)$. 

In the absence of a general analytic solution the model, the proper way to address this issue is through numerical simulation. However, economists have typically eshewed such methods in favor of analytic techniques for approximating the speed of convergence in the neighborhood of the long-run steady state. 

The most common anaytic technique for approximating $k(t)$ close to the long-run steady state is called [linearization](http://en.wikipedia.org/wiki/Linearization). The objective is to determine how rapidly $k$ approaches $k^{*}$. We know that $\dot{k}$ is determined by $k$:

$$\dot{k} = sf(k) - (n + g + \delta)k \tag{1.5.1}$$

which allows us to write $\dot{k}(k)$. Note that when $k=k^*$, $\dot{k}=0$. In order to derive an analytic expression for the speed of convergence using linearization we take a first-order Taylor approximation of $\dot{k}(k)$ around the point $k=k^*$.

$$ \dot{k} \approx \left[\frac{\partial \dot{k}(k)}{\partial k}\bigg|_{k=k^*}\right](k - k^*) \tag{1.5.2}$$

That is, $\dot{k}$ is roughly equal to the product of the difference between $k$ and $k^*$ and the derivative of $\dot{k}$ with respect to $k$ evaluated at $k=k^*$. If we define

$$\lambda = \left[\frac{\partial \dot{k}(k)}{\partial k}\bigg|_{k=k^*}\right], \tag{1.5.3}$$

then equation 1.5.2 becomes 

$$\dot{k} \approx \lambda(k(t) - k^*). \tag{1.5.4}$$

Equation 1.5.4 is a linear differential equation which can be solved using standard methods (see chapter 14 of Chiang and Wainwright's [*Fundamental methods of mathematical economics*](http://www.drchristiansalas.org.uk/BusinessandEconomics/Economics/FundamentalMethods.pdf) for the basics). The solution to equation 1.5.4 is

$$ k(t) = k^* + e^{\lambda t} (k_0 - k^*). \tag{1.5.5}$$

Note that in order for $k(t) \rightarrow k^*$ as $t\rightarrow\infty$ we require that $\lambda < 0$. For the Solow model it turns out that we can derive a general expression for $\lambda$.

\begin{align}
\lambda = \left[\frac{\partial \dot{k}(k)}{\partial k}\bigg|_{k=k^*}\right] =& sf'(k^*) - (n + g + \delta) \\
=&\frac{sf(k^*)f'(k^*)}{f(k^*)} - (n + g + \delta) \\
=&\frac{(n + g + \delta)k^*f'(k^*)}{f(k^*)} - (n + g + \delta) \\
=&\left(\frac{k^*f'(k^*)}{f(k^*)} - 1\right)(n + g + \delta) \\
=&-\left(1 - \alpha_k(k^*)\right)(n + g + \delta) \tag{1.5.6}
\end{align}

Similar arguments can be used to show that 

$$ y(t) = y^* + e^{\lambda t} (y_0 - y^*). \tag{1.5.7}$$

Substituting equations 1.5.5 and 1.5.6 into equation 1.5.4 yields the following analytic expression for the speed of convergence.

$$\dot{k} \approx -(1 - \alpha_k(k^*))(n + g + \delta)e^{-(1 - \alpha_k(k^*))(n+g+\delta) t}(k_0 - k^*) \tag{1.5.8}$$



## 1.6 Solow model and the central questions of growth theory

Central conclusion of the Solow model is that if the returns that capital commands in the market are a rough guide to its contribution to output, then variation in the accumulation pf physical capital do not account for a significant part of either worldwide economic growth or cross-country income differences.

Solow model implies that difference in output per worker in countries $i$ and $j$ (assuming common $\alpha_K$ and common $A$!) is... 

$$\ln\left(\frac{Y}{L}\right)_i - \ln\left(\frac{Y}{L}\right)_j = \alpha_K \left[\ln\left(\frac{K}{L}\right)_i - \ln\left(\frac{K}{L}\right)_j\right]$$

```python
def A_tilde(data, ctry, params):
    """Transform PWT TFP into TFP in a CES production function."""
    alpha = params['alpha']
    sigma = params['sigma']
    rho   = (sigma - 1) / sigma
    
    A = data.rtfpna[ctry]
    K = data.rkna[ctry]
    L = data.emp[ctry]
    H = data.hc[ctry]
    
    TFP = (K / L)**alpha * (((A * H**(1 - alpha))**rho - alpha) / (1 - alpha))**(1 / rho)
    
    return TFP
```

```python
A_tilde(pwt_data, 'GBR', {'alpha':0.33, 'sigma':1.5})
```

```python
def func(params, data, ctry):
    """optimize to find alpha and sigma."""
    params = {'alpha':params[0], 'sigma':params[1]}
    
    alpha = params['alpha']
    sigma = params['sigma']
    rho   = (sigma - 1) / sigma
    
    Y = data.rgdpna[ctry]
    K = data.rkna[ctry]
    L = data.emp[ctry]
    A = A_tilde(data, ctry, params)
    
    y = (Y / (A * L))
    k = (K / (A * L))
    
    # remove the discontinuity 
    if abs(rho) < 1e-6:
        return np.sum((np.log(y) - alpha * alpha * np.log(k))**2)
    else:
        return np.sum((np.log(y) - (1 / rho) * np.log(alpha * k**rho + (1 - alpha)))**2)
```
