##Task 1: Coding a stochastic version of the Solow model in Python

The following are the equations describing our stochastic Solow model:

* Intensive form of the production function: $y_{t} = f(k_{t})$
* A mean zero distribution of technology innovations: $\epsilon \sim N(0, \sigma_{A})$
* Persistent deviations of technology from trend: $\ln\ z_{A,t+1} = \rho \ln\ z_{A,t} + \epsilon_{t+1},\ |\rho| \lt 1$ 
* Technology process: $A_{t+1} = (1 + g)\frac{z_{A,t+1}}{z_{A,t}}A_{t}$
* Evolution of capital: $k_{t+1} = \left(\frac{1}{(1+g)(1+n)}\right)\left(\frac{z_{A,t}}{z_{A,t+1}}\right)\left[sf(k_{t}) + (1 - \delta)k_t\right]$

By now you should be pretty comfortable writing Python functions, so in this lab we expect more from you in terms of writing your own code.  You are given some of the functions that you will need to complete the lab, but are expected to code others yourself.
 

```python
def f(k):
    """In your lectures you will have been introduced to the intensive form 
    of the production function.  The intensive form of the production 
    function expresses output per effective worker, y, as a function of 
    the capital stock per effective worker, k.
    
    Inputs:
        1) k: capital per effective worker
        
    Returns:
        1) y: output per effective worker
        
    """
    # extract params
    alpha = original_params['alpha']
    
    return k**alpha
    
def capital(k, z, zplus):
    """Function that takes end of period t's capital stock per 
    effective worker, k_t, and a contemporaneous technology 
    shock, z_{t+1}, and returns the end of period t+1's capital 
    stock per effective worker, kplus.

    k_{t+1} = \\frac{1}{(1 + g)(1 + n)}\\frac{z_t}{z_{t+1}}[(1 - \delta)k_t + sk_t^{\alpha}]
        
    Inputs: 

        1. capital per effective worker, k.
        2. technology shock, zplus (default = 1).

    Returns: next period's value of capital per effective worker,
             kplus.
    
    """
    # extract params
    s     = original_params['s']
    n     = original_params['n']
    g     = original_params['g']
    alpha = original_params['alpha']
    delta = original_params['delta']
    
    # next period's capital per effective worker
    kplus = (1 / ((1 + g) * (1 + n))) * (z / zplus) * ((1 - delta) * k + s * k**alpha) 

    return kplus

def techShock(z, eplus):
    """Defines a stochastic process representing technology shocks.

    Inputs:

        1. z: previous value of the technology shock, z.
        2. e: a draw from a N(0, sigma) RV.

    Returns: current value for the technology shock.
        
    """    
    # extract params
    rho = original_params['rho']

    # technology shock is log-normal
    zplus = z**rho * np.exp(eplus)
    return zplus
    
def technology(a, z, zplus):
    """Evolution of technology.
    
        A_{t+1} = (1 + g)(\frac{z_{A,t+1}}{z_{A,t}})A_t

    Inputs:
        - current technology, a
        - current technology deviation, zplus
         
    Returns:
        - next period's technology, aplus
   
    """
    # extract params
    g = original_params['g']
    aplus = (1 + g) * (zplus / z) * a
    
    return aplus  

def eps():
    """Mean zero Gaussian disturbance term"""
    # extract params
    sigma = original_params['sigma']

    # construct the shock using SciPy distribution
    shock = stats.norm(0, sigma)
    
    return shock.rvs(1)
```

###Exercise:

As in the deterministic version of the Solow growth model technology, A, grows at an exogenous rate $g$; however, in the stochastic version, the trend growth rate g is subject to persistent random shocks.

$$A_{t+1} = (1 + g)\left(\frac{z_{A,t+1}}{z_{A,t}}\right)A_{t}$$

where the technology shocks in the stochastic Solow model (as well as most other models you will encounter) are assumed to follow a first-order autoregressive, AR(1), process: 

$$ln\ z_{A,t+1} = \rho ln\ z_{A,t} + \epsilon_{t+1}$$

The persistence of the AR(1) process is governed by the parameter $-1 \lt \rho \lt 1$. The disturbance term epsilon is assumed to be $N(0, \sigma)$ for all $t$. Note that setting $\rho = 0$ implies that technology shocks are independent. In the literature you will often see these shocks referred to as either 'productivity shocks' or 'technology shocks.'

##Task 3: Understanding the difference between deterministic and stochastic steady states.

Unlike the deterministic Solow model, in the stochastic Solow model the steady state is no longer a single value of $k$, but instead an entire distribution of $k$. This steady state distribution is sometimes also referred to in the economics literature as the ergodic/stationary/invariant distribution. We will estimate the steady state distribution of $k$ and $c$ via simulation below.  In more sophisticated stochastic models with optimizing agents, like the RBC model which we will cover in detail next semester, we will need to approximate the optimal decision rules around some point in the model's phase space.  But which point?  In a deterministic model, such as the Ramsey model from the previous lab, there was only a single option: the unique deterministic steady state.  However, in a stochastic model there are potentially many points around which we could approximate decision rules.  This task is designed to demonstrate the difference between various notions of steady state in a stochastic environment.

```python
def freedman_diaconis_rule(data, default=10):
    """Implements the Freedman-Diaconis rule for optimally choosing the bin 
    size for a histogram and returns optimal number of bins.
        
    """
    # calculate IQR
    upper_quartile = stats.scoreatpercentile(data, 75)
    lower_quartile = stats.scoreatpercentile(data, 25)
    iqr = upper_quartile - lower_quartile
        
    # calculate optimal bin width
    h = 2 * (iqr / (data.size)**(1. / 3.))
        
    # calculate number of bins (making sure result is an int)
    if h != 0:
        bins = int((np.max(data) - np.min(data)) / h)
    else:
        bins = default
        
    return bins
```

```python
# simulate the a long time series of data
data = model.get_samplePath('solow', T=1e6)
```

```python
# Plot a histogram of the stationary distribution
plt.figure(figsize=(8,8))

# choose the optimal number of bins...
bins = freedman_diaconis_rule(data[:, 0])

# plot the histogram
plt.hist(data[:, 0], bins=bins, normed=True, alpha=0.75, edgecolor='w')

# Add the deterministic steady state level of capital stock per effective worker
plt.axvline(model.SS_dict['k_star'], ls='--', color='k', label=r'$k_D^{*}$')

# demarcate the mean of the steady state distribution of k
plt.axvline(np.mean(data[:, 0]), ls='-.', color='k', linewidth=3, label=r'$E[k]$')

# Don't forget to label your axes!
plt.xlabel('Capital per effective worker, k')
plt.ylabel('Density')

# Adding a title
plt.title('Ergodic distribution of capital per effective worker, $k$', 
          weight='bold')

# Add legend
plt.legend(loc='best', frameon=False)

#plt.savefig('Graphics/Ergodic-Distribution-k.png')
plt.show()
```
